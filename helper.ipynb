```python
losses = []
val_losses = []
test_losses = []
with open("logfile.txt", 'r') as inputfile:
    for line in inputfile:
        if line[0]=="T":
            losses.append([float(item) for item in line[2:].split(",")])
        elif line[0]=="V":
            val_losses.append([float(item) for item in line[2:].split(",")])
        else:
            test_losses.append([float(item) for item in line[2:].split(",")])
losses = np.asarray(losses)
val_losses = np.asarray(val_losses)
test_losses = np.asarray(test_losses)
```

```python
print(losses.shape)
print(val_losses.shape)
print(test_losses.shape)
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(131)
for i in range(1, 8):
    plt.plot(losses[10:, 0], losses[10:, i])
plt.title("Deep Feature Loss Train")
plt.subplot(132)
for i in range(1, 8):
    plt.plot(val_losses[:, 0], val_losses[:, i]/1e9)
plt.title("Deep Feature Loss Val")
plt.subplot(133)
for i in range(1, 8):
    plt.plot(test_losses[:, 0], test_losses[:, i]/1e9)
plt.title("Deep Feature Loss Test")
```
